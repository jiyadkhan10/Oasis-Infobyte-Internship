# Data Cleaning

```python
data.rename(columns = {'Region.1':'Direction'}, inplace = True) 
```

```python
data
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

## ðŸŒŸRegion

```python
data.loc[data['Region'].isna()]
```

```python
data.dropna(how='all', inplace=True)
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

## ðŸŒŸ Area

```python
data.loc[data['Area'].isna()].reset_index()
```

```python
data['Area']=data.groupby('Region')['Area'].transform(lambda x: x.fillna(x.mode().iloc[0]))
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

## ðŸŒŸDirection

```python
data['Direction'].value_counts().reset_index()
```

```python
data.loc[data['Direction'].isna()].reset_index()
```

```python
data['Direction'] = data.groupby('Region')['Direction'].transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

```python
data[data['Direction'].isna()]
```

```python
data.fillna(value={'Direction': "Northwest"}, inplace=True)
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

## ðŸŒŸLongitude

```python
data.loc[data['longitude'].isna()].reset_index()
```

```python
data['longitude'] = data.groupby('Region')['longitude'].transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

```python
data[data['longitude'].isna()]
```

```python
data.fillna(value={'longitude': 30.7333}, inplace=True)
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

## ðŸŒŸLatitude

```python
data.loc[data['latitude'].isna()].reset_index()
```

```python
data['latitude'] = data.groupby('Region')['latitude'].transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

```python
data[data['latitude'].isna()]
```

```python
data.fillna(value={'latitude': 76.7794}, inplace=True)
```

```python
data.isna().sum().reset_index().rename(columns={0:'count'})
```

# Data Visualization

```python
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
pio.renderers.default = 'notebook'
import plotly.offline as pyo
pyo.init_notebook_mode(connected=True)
pd.options.display.max_columns=100
```

```python
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
corr = numeric_columns.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(corr, annot=True, cmap=sns.cubehelix_palette(as_cmap=True), fmt=".2f", linewidths=0.5)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
data_2019 = data[data['Year'] == 2019]
month_order = [calendar.month_name[i] for i in range(1, 13)]
data_2019['Month'] = pd.Categorical(data_2019['Month'], categories = month_order, ordered = True)


data_2019_monthly_avg = data_2019.groupby('Month')['Estimated Unemployment Rate'].mean().reset_index()

figure_2019 = px.line(data_2019_monthly_avg, x = 'Month', y = 'Estimated Unemployment Rate', title = 'Unemployment Rate Trend (2019)',
                   labels = {'Estimated Unemployment Rate': 'Unemployment Rate (%)'},
                   line_dash_sequence = ['solid'], color_discrete_sequence = ['yellow'])

trace_2019 = go.Scatter(x = data_2019_monthly_avg['Month'], y = data_2019_monthly_avg['Estimated Unemployment Rate'], 
                        mode = 'markers+lines', name = '2019', line = dict(color='yellow'))

figure_2019.add_trace(trace_2019)
figure_2019.update_layout(
    xaxis_title = 'Month',
    yaxis_title = 'Unemployment Rate (%)',
    template = 'plotly_white'
)

figure_2019.show()
```

```python
data_2020 = data[data['Year'] == 2020]
month_order = [calendar.month_name[i] for i in range(1, 13)]
data_2020['Month'] = pd.Categorical(data_2020['Month'], categories = month_order, ordered = True)


data_2020_monthly_avg = data_2020.groupby('Month')['Estimated Unemployment Rate'].mean().reset_index()

figure_2020 = px.line(data_2020_monthly_avg, x = 'Month', y = 'Estimated Unemployment Rate', title = 'Unemployment Rate Trend (2020)',
                   labels = {'Estimated Unemployment Rate': 'Unemployment Rate (%)'},
                   line_dash_sequence = ['solid'], color_discrete_sequence = ['red'])

trace_2020 = go.Scatter(x = data_2020_monthly_avg['Month'], y = data_2019_monthly_avg['Estimated Unemployment Rate'], 
                        mode = 'markers+lines', name = '2020', line = dict(color='red'))

figure_2020.add_trace(trace_2020)
figure_2020.update_layout(
    xaxis_title = 'Month',
    yaxis_title = 'Unemployment Rate (%)',
    template = 'plotly_white'
)

figure_2020.show()
```

```python
data_2019 = data[data['Year'] == 2019]
data_2020 = data[data['Year'] == 2020]

avg_unemployment_2019 = data_2019.groupby('Region')['Estimated Unemployment Rate'].mean().reset_index()
avg_unemployment_2020 = data_2020.groupby('Region')['Estimated Unemployment Rate'].mean().reset_index()

combined_data = pd.merge(avg_unemployment_2019, avg_unemployment_2020, on = 'Region', suffixes = ('_2019', '_2020'))

figure = px.bar(combined_data, x = 'Region', y = ['Estimated Unemployment Rate_2019', 'Estimated Unemployment Rate_2020'],
             barmode = 'group', labels = {'value': 'Unemployment Rate (%)'},
             title = 'Average Estimated Unemployment Rate by Region (2019 vs. 2020)')

figure.update_layout(
    xaxis_title = 'Region',
    yaxis_title = 'Unemployment Rate (%)',
    template = 'plotly_white'
)

figure.show()
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

data['Month'] = pd.Categorical(data['Month'], categories = month_order, ordered = True)
data_2019 = data[data['Year'] == 2019]
data_2020 = data[data['Year'] == 2020]

avg_unemployment_2019 = data_2019.groupby(['Month', 'Area'])['Estimated Unemployment Rate'].mean().reset_index()
avg_unemployment_2020 = data_2020.groupby(['Month', 'Area'])['Estimated Unemployment Rate'].mean().reset_index()

avg_unemployment_2019['Year'] = 2019
avg_unemployment_2020['Year'] = 2020

combined_data = pd.concat([avg_unemployment_2019, avg_unemployment_2020], ignore_index = True)

figure = px.bar(combined_data, x = 'Month', y = 'Estimated Unemployment Rate', color = 'Area', facet_col = 'Year',
             labels = {'Estimated Unemployment Rate': 'Unemployment Rate (%)'},
             title = 'Unemployment Rate Comparison (2019 vs. 2020) by Area',
             color_discrete_map = {'Rural': 'blue', 'Urban': 'red'})

figure.update_layout(
    xaxis_title = 'Month',
    yaxis_title = 'Unemployment Rate (%)',
    template = 'plotly_white'
)

figure.show()
```
